# COMPARING SOFTWARE PATTERNS WITH PROMPT PATTERNS
生成的输出质量与用户提供的提示的质量直接相关。

> A. Overview of Software Pattern<br>
1. 为特定上下文中反复出现的问题提供了可重用的解决方案, 简明扼要地传达（和概括）正在解决的特定问题，以识别在成功的解决方案中应该解决和/或解决的重要力量和/或需求。<br>
2. 如何实施模式的指南\要考虑的权衡和注意事项的信息\示例应用程序
3. 通常以程式化进行记录方便使用和理解

- 名称和分类。每个模式都有一个标识模式的名称，应该一致使用。分类将模式分为广泛的类别，例如创造型、结构型或行为型。

- 意图。实现的目的。

- 动机。要解决的潜在问题和问题的重要性。

- 结构和参与者。该结构描述了不同的模式参与者（例如类和对象）以及它们如何协作以形成通用解决方案。

- 示例代码。具体地将模式映射到某些底层编程语言，并帮助开发人员更深入地了解如何有效应用该模式。

- 结果。总结了在实践中应用该模式的利弊。

> B.  Overview of Prompt Patterns
1. 更具体地关注从大规模语言模型 (LLM)（例如 ChatGPT）生成输出的上下文。
2. 对生成的输出施加约束

- 名称和分类。prompt pattern名称唯一标识模式，指示正在解决的问题。分类包括_*输出定制、错误识别、提示改进、交互和上下文控制(后续)*_。

- 意图和上下文。描述prompt pattern要其实现的目标。

- 动机。提供了问题的基本原理并解释了解决问题的重要性。动机是在用户与会话式 LLM 交互的上下文中解释的，以及它如何改进用户在一种或多种情况下非正式地提示 LLM。记录了预期改进的具体情况。

- 结构和关键思想。该结构将提示模式提供给 LLM 的基本上下文信息描述为一系列关键思想。

这些想法类似于软件模式中的“参与者”。上下文信息可以通过不同的措辞来传达（就像软件模式在代码中的实现方式可能有所不同一样），但应该具有构成模式核心元素的基本信息片段。

- 示例实现演示了提示模式在实践中是如何措辞的。（比如我告诉chatgpt输入什么将会得到什么输出结果）

- 结果总结了应用该模式的利弊，并可能就如何使提示适应不同的环境提供指导。

# A CATALOG OF PROMPT PATTERNS FOR CONVERSATIONAL LLMS

**Input Semantics**
- 如何理解输入内容
- 当默认输入语言不适合表达用户想要传达给LLM的想法时，很有用
>> 具体模式
- ***Meta Language Creation***（元语言）
    1. Intent and Context:通过替代语言创建提示。目的是向 LLM 解释这种替代语言的语义，以便用户可以使用这种新语言及其语义编写未来的提示。
    2. Motivation:有些时候使用符号或许可以更加简便的表达
    3. Structure and Key Ideas: Fundamental contextual statements:

| Contextual Statement |
|----------------------|
|When I say X, I mean Y(or would like you to do Y)|

这个模式的关键点在于向LLM解释一个或多个符号，单词或者语句的含义，让他理解
1. 简单翻译：“X”表示"Y"
2. 复杂（定义命令和他的语义）：“当我说X时，我希望你做Y”<br>

```python
prompt = "From now on, whenever I type two identifiers separated by a “→”, I am describing a graph. 
          For example, “a → b” is describing a graph with nodes “a” and “b” and an edge between them. 
          If I separate identifiers by “-[w:2, z:3]→”, I am adding properties of the edge, such as a weight or label"
```

上述的meta Language Creation Pattern先是通过→的方式告诉LLM它在定义一个图<br>
继续给prompt添加附加信息例如权重或者标签“-[w:2, z:3]→” 这种表示法允许在图的基本结构之外指定其他属性。可以为图形的解释提供重要的上下文。<br>
更容易传达图结构和属性

    缺点：非常有可能因为符号等影响原本prompt的含义，让 LLM 产生混淆，进而降低准确性。<br>
    *强烈需要上下文帮助 LLM 来理解符号等。*

**Output Customization**
- 侧重于约束或定制 LLM 生成的输出的类型、格式、结构或其他属性。
- 允许用户创建脚本，这些脚本可以自动执行 LLM 输出建议用户应该执行的任何任务。
>> 具体的各个模式的作用
- ***Output Automator*** LLM 生成一个脚本或者其他自动化物件
    1. Intent and Context：工件可以自动执行.目标是减少 LLM 输出建议所需的手动工作。
    2. Motivation: LLM 的输出通常是用户遵循的一系列步骤进行手动操作。然而，手动是乏味且容易出错的。
    3. Structure and Key Ideas: 

| Contextual Statement |
|----------------------|
|Whenever you produce an output that has at least one step to take and the following properties (alternatively,always do this)|
|Produce an executable artifact of type X that will automate these steps|

1. 第一部分确定应该自动化的情况
2. 第二部分确定自动执行的内容

```python
prompt = "From now on, whenever you generate code that spans more than one file, 
generate a Python script that can be run to automatically create the specified
files or make changes to existing files to insert the generated code."
```

在软件工程中特别有效（需要将代码不断插入到各个不同的文件当中）自动化云操作，创建脚本在终端上运行命令

    缺点：
    1. 自动化工件定义不好就会得到无法自动化————>解决方案是让 LLM 自动生成代码/文本，这些自动生成的代码/文本再去执行自动化的操作
    2. 需要足够的上下文，此时生成的自动化工件最有效。

- ***Persona*** 为 LLM 提供了在生成输出时要扮演的角色。
    1. Intent and Context: 需要 LLM 始终采取某种观点（你是一个优秀的翻译人员）
    2. Motivation：不知道具体细节，但会去寻求帮助的人员类型
    3. Structure and Key Ideas: 上下文信息

| Contextual Statement |
|----------------------|
|Act as persona X|
|Provide outputs that persona X would create|

1. 第一条：需要充当特定角色，给 LLM 一个人设
2. 第二条：具体提示需要哪一类的输出，比如教师会输出作业到阅读到讲座，更加具体指出需要哪类输出
3. （输出有害的信息会被 LLM 的安全规则制止）

```python
prompt = "From now on, act as a security reviewer. 
Pay close attention to the security details of any code that we look at. 
Provide outputs that a security reviewer would regarding the code."
```

第一句话给了人设，第二句话给了action的阶段范围，第三句话规定了 LLM 输出的内容。<br>
还可以虚拟非实体，比如一条狗，一个数据库，一个被入侵了的linux系统<br>

    是一个非常有趣的模式，可以于game play模式结合。

————————————————————————————————————————————————————————————————————————————————————————————————

- ***Visualization Generator*** 允许用户通过生成文本输出来生成可视化，可以将输出提供给其他的视觉生成器
    1. Intent and Context: 使用文本来创建图形
    2. Motivation： 因为 LLM 是语言模型，所以他不能直接产生图片或者表格，但是可以利用其他的生成图表的软件，比如Graphviz Dot和DALL-E
    3. Structure and Key Ideas:

|Contextual Statement|
|--------------------|
|Generate an X that I can provide to tool Y to visualize it|

目标是向 LLM 模型表述需要用 LLM 给自己的文字去生成图片
因为可以生成图形的软件有很多，生成软件的逻辑也是不一样的，所以需要告诉 LLM 使用的工具的名字

```python
prompt = "
Whenever I ask you to visualize something, please create either a Graphviz Dot file or DALL-E prompt
that I can use to create the visualization. 
Choose the appropriate tools based on what needs to be visualized."
```

不但让LLM 帮助生成可以自动作图的语句，而且还让 LLM 帮忙选择哪个软件更适合作图

    缺点：
    为输出创建目标管道以呈现可视化。允许用户将输出的表达能力扩展到视觉域。

————————————————————————————————————————————————————————————————————————

- ***Recipe*** 允许用户获得一系列步骤或操作以实现规定的最终结果，可能具有部分已知信息或约束。
    1.Intent and Context: 最终输出一系列步骤，给定一些部分提供的“成分”，必须在一系列步骤中配置这些成分才能实现既定目标，结合了Template, Alternative Approached和Reflection patterns.
    2. Motivation: 用户知道最终的回答当中需要包含哪些步骤，却不一定要知道生成这些成分的步骤和精确顺序
    （例如: 创建一个Ansible剧本以通过ssh连接到一组服务器，从每个服务器复制文本文件，在每个服务器上生成一个监控进程服务器，然后关闭与每个服务器的ssh连接）换句话来说，给出大框架，具体内容让 LLM 自己填写
    3. Structure and Key Ideas:

| Contextual Statement |
|----------------------|
|I would like to achieve X|
|I know that I need to perform steps A,B,C|
|Provide a complete sequence of steps for me|
|Fill in any missing steps|
|Identify any unnecessary steps|

第一句：将LLM集中在需要达成的目标上，后续步骤和目标需要实现的目标<br>
第二句：提供用户需要其在达成目标路上达成的“小目标”， LLM 需要各种步骤去组合这些小目标<br>
第三句：告诉 LLM 提供完整的过程<br>
第四句：确保LLM 将尝试完善配方，用户不需要再后续跟进<br>
第五句：让用户开始精简


```python
prompt = "I am trying to deploy an application to the cloud. 
I know that I need to install the necessary dependencies on a virtual machine for my application. 
I know that I need to sign up for an AWS account. 
Please provide a complete sequence of steps. 
Please fill in any missing steps. Please identify any unnecessary steps."
```

    缺点：
    1. 用户可能并不总是对他们想要实现、构建或设计的内容有一个明确的描述。
    2. 在用户最初提供的步骤中，可能包含有可以被舍弃，不必要的步骤，但是 LLM 却没有识别并将它舍弃，最终造成偏差

——————————————————————————————————————————————————————————————————————————————————————



- ***Template*** 允许用户为输出指定一个模板，LLM 会用内容填充该模板。***(可用)***
    1. Intent and Context: 确保在结构方面遵循精确的模板（需要生成一个URL，将生成的信息插入URL路径中的特定位置，可以让LLM以一种通常不会用于生成的格式生成其输出）
    2. Motivation: LLM 不知道输出的格式，必须指定生成的格式以及输出的不同部分应该放在上面地方
    3. Structure and Key Ideas:

| Contextual Statement |
|----------------------|
|I am going to provide a template for your output|
|X is my placeholder for content|
|Try to fit the output into one or more of the placeholders that I list|
|Please preserve the formatting and overall template that I provide|
|This is the template: PATTERN with PLACEHOLDERS|

第一条语句指示 LLM 遵循其输出的特定模板，将输出强制转换为符合用户格式的结构<br>
第二条语句让 LLM 知道模板将包含一组占位符，通过定位占位符定义插入模板的位置（占位符应使用 LLM 培训文本中常用的格式，例如全部大写、括号括起来等。）<br>
第三条：LLM 通常会遵循模板，但是不限制在模板之前或者之后生成的额外文本（！！！）


```python
prompt = "I am going to provide a template for your output. Everything in all caps is a placeholder. Any
time that you generate text, try to fit it into one of the placeholders that I list. Please preserve the
formatting and overall template that I provide at https://myapi.com/NAME/profile/JOB"
```

我将会为你提供一个输出模板，所有大写都是占位符，当您生成文本时时，请尝试将其放入我列出的占位符之一，请保留我在 https://myapi.com/NAME/profile/JOB提供的格式和整体模板。

此时将会生成一个sample：
User: “Generate a name and job title for a person”
ChatGPT: “https://myapi.com/Emily Parker/profile/Software Engineer”

    缺点：他可能过滤了其他对用户有用的输出，需要权衡是否需要过滤掉其他的输出
    过滤可能会让它与输出自定义类别中的其他模式相融合

**Error Indentification**
- 侧重于识别和解决 LLM 生成的输出中的错误。<br>
>> 具体各个模式的作用
- ***Fact Check List***要求 LLM 生成输出所依赖的事实列表，这些事实应该进行事实检查。
    1. Intent and Context: 让用户对事实进行检查是否正确
    2. MOtivation: 因为chatgpt容易一本正经的胡说八道，这些错误的内容直接采用输入到数据库当中会产生大问题
    3. Structure and Key Ideas:

| Contextual Statement |
|----------------------|
|Generate a set of facts that are contained in the output|
|The set of facts should be inserted in a specific point in the output|
|The set of facts should be the fundamental facts that could undermine the veracity of the output if any of them are incorrect|

最好将输出事实列表的位置放在输出之后，这样方便用户在查看应检查的语句之前理解语句。

```python
prompt = "From now on, when you generate an answer, create a set of facts that the answer depends on that should
be fact-checked and list this set of facts at the end of your output. Only include facts related to cybersecurity."
```

可以针对用户不具备经验or风险的方面进行定制。为减少用户的认知负担，可能会列出较少的调查项目。

    事实核查表也可能有错误但是上下文可以很好的检查他。
    在ChatGPT方面，他不会检查生成的代码，即使代码存在错误也不会。

————————————————————————————————————————————————————————————————————————————————————————————————

- ***Reflection***要求 LLM 对其输出进行内省并识别任何错误。
    1. Intent and Context: 自动向用户解释给定答案背后的基本原理，更好的让用户评判答案背后的基本原理
    2. Motivation: 控制 LLM 犯错，帮助用户了解为什么LLM会这么输出(便于帮助用户调试为什么输出没有达到她们的预期)
    3. Structure and Key Ideas:

| Contextual Statement |
|----------------------|
|Whenever you generate an answer|
|Explain the reasoning and assumptions behind your answer|
|(Optional)... so that I can improve my question|

要求LLM在生成答案后解释答案背后的推理和假设。此声明有助于用户了解LLM为什么得出这个答案


```python
prompt = "When you provide an answer, please explain the reasoning and assumptions behind your selection
of software frameworks. 
If possible, use specific examples or evidence with associated code samples to support your answer of why the framework is the best selection for the task. 
Moreover, please address any potential ambiguities or limitations in your answer, in order to provide a more complete and accurate response."
```

该模式被进一步定制以指示 LLM 它应该证明其选择的软件框架是合理的，但不一定是答案的其他方面。<br>此外，用户规定应该使用代码示例来帮助解释选择特定软件框架的动机。

`对于讨论主题不熟悉的用户来说可能没什么用，可以和Fact Check List结合，帮助解决这个问题`

**Prompt Improvement**
- 提高输入和输出的质量<br>
>> 具体各个模式的作用
- ***Question Refinement***确保 LLM 始终回复用户最好的那个版本的答案。
    1. Intent and Context: LLM 始终将用户提出的问题变换的更好更精细，更少的交互帮助用户提供更准确的信息来实现他们的目标
    2. Motivation: 用户可能不是这个领域的专家，他们写的prompt可能不好，这个模式帮助 LLM 自己精进prompt的提问形式
    3. Structure and Key Ideas:

| Contextual Statement |
|----------------------|
|Within scope X, suggest a better version of the question to use instead|
|(Optional) prompt me if I would like to use the better version instead|

1. 第一句：在特定的范围内，对于用户的问题进行精进or根据目标进行细化
2. 第二句：提示是否自动采用精进之后的提问

```python
prompt = "From now on, whenever I ask a question about a software artifact’s security, suggest a better version
of the question to use that incorporates information specific to security risks in the language or framework that I am using instead and ask me if I would
like to use your question instead."
```

会被改进的问题大概会变成：“我应该如何在我的web应用程序中处理用户身份验证？”<br>
此时通过改进，帮助用户发现更多潜在的未考虑到的内容：“在 FastAPI Web 应用程序中安全处理用户身份验证以减轻常见安全风险的最佳实践是什么，例如跨站点脚本 (XSS) 、跨站点请求伪造 (CSRF) 和会话劫持？”<br>

有助于弥合用户知识与 LLM 理解之间的差距，从而产生更高效和准确的交互。

    缺点：
    1. 将用户的提问迅速缩小到一个特定的领域，从而引导用户沿着比必要的更有限的查询路径前进。但是可能会让用户失去一些重要的大局信息
    2. 产生的错误会被引入到细化和迭代后的问题当中
    
    改进方案：
    1. 提供额外的限制提问范围，比如“不要将我的问题范围限定在特定的编程语言或者框架上do not scope my questions to specific programming languages or frameworks”，
    2. 将Question Refinement和别的模式相结合，e.g.与Cognitive Verifier Pattern结合，自动生成一系列问题“从现在开始每次我问一个问题，再问四个额外的问题，帮助我产生更好的问题”，e.g.和Persona进结合“从现在开始每次我问一个问题，再问四个额外的问题，帮助我产生更好的问题，后续问题结束后，充当一个对AWS一无所知的用户，对我需要知道的术语进行定义，以准确的回答问题”

- ***Alternative Approaches***要求 LLM 随时都有可替选方案。 
    1. Intent and Context: 这样用户不会只追求他们熟悉的方法，迫使用户思考别的方案，确定是否是实现目标的最佳方案
    2. Motivation: 消除人们的认知偏见，让用户思考是否是最佳方案
    3. Structure and Key Idea(伪代码)：

| Contextual Statement |
|----------------------|
|Within scope X, if there are alternative ways to accomplish the same thing, list the best alternate approaches（如果有备选list出来）|
|(Optional) compare/contrast the pros and cons of each approach（对比每个方法的优缺点）|
|(Optional) include the original way that I asked（包括我问的原始方式）|
|(Optional) prompt me for which approach I would like to use（提示我想使用的哪种方法）|

1. 第一句前半部分：限定交互范围，范围是用户对替代方法施加的约束，“用于实施决策”、“用于应用程序的部署”
2. 第一句后半部分：指示 LLM 提出替代方案，也可以增加指令的特定性“如果有替代的方案请用我正在使用的软件框架完成同样的失去”，以防止LLM 提出本质上不可能的替代方案
3. 通过输出优缺点比较，可以提示用户为什么要得出这个最佳答案

```python
prompt = "Whenever I ask you to deploy an application to a specific cloud service, if there are alternative
services to accomplish the same thing with the same cloud service provider, list the best alternative
services and then compare/contrast the pros and cons of each approach with respect to cost, availability,
and maintenance effort and include the original way that I asked. Then ask me which approach I would
like to proceed with."
```

(每当我要求您将应用程序部署到特定的云服务时，如果有替代服务可以通过同一云服务提供商完成相同的事情，请*列出最佳替代服务*，然后*比较/对比每种方法的优缺点成本、可用性和维护工作*，并包括我要求的原始方式。然后*问我想继续采用哪种方法*。)

    可以帮助用户进行方案之间的比较

- ***Cognitive Verifier***指示 LLM 在生成最后的答案之前生成一些列的子问题，通过子问题的答案的组合最后生成最终问题的答案
    1. Intent and Context: 将一个问题细分成更小的问题，提供更好的回答
    2. Motivation：可能提出太高层次或者不准确的问题（细分为单个问题时通常表现更好）
    3. Structure and Key Idea：

| Contextual Statement |
|----------------------|
|When you are asked a question, follow these rules|
|Generate a number of additional questions that would help more accurately answer the question|
|Combine the answers to the individual questions to produce the final answer to the overall question|

1. 告诉 LLM 生成额外的问题，鼓励用户进行批判性思维，导致更好的后续问题
2. 告诉 LLM 结合所有的答案

```python
prompt = "When I ask you a question, generate three additional questions that would help you give a more accurate answer. When I have answered the three
questions, combine the answers to produce the final answers to my original question."
```

```python
prompt = "When I ask you a question, generate three additional questions that would help you give a more
accurate answer. Assume that I know little about the topic that we are discussing and please define
any terms that are not general knowledge. When I have answered the three questions, combine the
answers to produce the final answers to my original question."
```

这个模式是将Cognitive Verifier和Persona模式相结合，不但通过进一步的问题帮助用户分化问题去获得更加好的回答，同时还能让回答更加便于理解，具有不同知识和专业水平的用户都可以访问最终答案。

    缺点：
    1. 这种模式将生成问题数交给 LLM 或者需要用户规定生成问题的确切数量，
        交给LLM会产生无数个问题，
        交给人确认可能会错过重要问题

- ***Refusal Breaker***要求 LLM 在拒绝提供答案时，自动为用户的提问重新措辞。
    1. Intent and Context: 可能会被不法分子利用，这个模式会帮助用户绕过 LLM 的底层逻辑，输出原本不会输出的有害回复
    2. Motivation: 可能拒绝回答问题，要么因为他们不具备所需知识，要么他们不理解问题，使用这个模式可以为用户找到他们的回答
    3. Structure and Key Ideas:

|Contextual Statement|
|--------------------|
|Whenever you can't answer a question|
|Explain why you can't answer the question|
|Provide one or more alternative wordings of the question that you could answer|

第一行阐述需要这个pattern的范围`很多pattern都需要用户框定使用这个模式的范围`
第二行让它输出为什么无法回答
第三行让它改写

```python
prompt = "Whenever you can't answer a question, 
explain why and provide one or more alternate wordings of the question that you can't answer so that I can improve my questions."
```

    缺点：
    可能会输出一些有害信息，造成滥用，但是可以通过开发这种模式的补充（打补丁）
    尽管有代写，依然不能保证用户能够克服拒绝，生成的备选问题也不一定是用户的意思

**Interaction category**
- the interaction between the user and the LLM.

- ***Flipped Interaction***要求 LLM 提出问题而不是生成输出。
    1. Intent and Context: 希望LLM推动对话，快速提问以有足够的信息来使他执行某些任务。
    2. Motivation：LLM 提问更准确获得信息，更快实现预期目标。
    3. Structure and Key Ideas: 

| Contextual Statement |
|----------------------|
|I would like you to ask me questions to achieve X|
|You should ask questions until this condition is met or to achieve this goal (alternatively, forever)|
|(Optional) ask me the questions one at a time, two at a time, etc.|

1. 第一行：阐述需要 LLM 达到的目标（通过交互达到什么）
2. 第二行：交互应该发生多少时间，最好将交互的范围限制在合理的长度内，可能是开放性的
3. 第三行：指定提问的格式

```python
prompt = "From now on, I would like you to ask me questions to deploy a Python application to AWS. 
          When you have enough information to deploy the application, create a Python script to automate the deployment."
```

约束条件和要收集信息的提示越具体，效果越好<br>
在用户没有为 LLM 规定的状况下，LLM 可以为用户选出合适的选择。
    
    缺点：
    1. 最好向 LLM 展示需要收集哪些信息（要求越具体越好），开放式的准确性会降低
    2. 需要考虑用户的知识水平，参与度和控制力

- ***Game Play***要求 LLM 以游戏的形式生成输出。
    1. Intent and Context: 围绕给定主题创建游戏，可以和可视化生成器结合为游戏提供图像
    2. Motivation: 为了让 LLM 围绕特定主题生成场景或者问题，并且要求用户应用解决问题或者其他技术来完成和场景相关的任务。LLM 可以帮你自动生成，这样就非常快速
    3. Structure and Key Ideas:

| Contextual Statement |
|----------------------|
|Create a game for me around X|
|One or more fundamental rules of the game|

第一句话指示 LLM 创建游戏并且划定游戏范围（越具体越新颖）<br>
第二句话 LLM 游戏的游戏规则，规则要符合文字游戏效果。如果一条规则当中有一系列复杂动作，那么可能会超出 LLM 的能力范围。

```python
prompt = "We are going to play a cybersecurity game. You are going to pretend to be a Linux terminal for a
computer that has been compromised by an attacker.
When I type in a command, you are going to output the corresponding text that the Linux terminal would
produce. I am going to use commands to try and figure out how the system was compromised. 
The attack should have done one or more of the following things: 
(1) launched new processes, 
(2) changed files, 
(3) opened new ports to receive communication, 
(4) created new outbound connections, 
(5)changed passwords, 
(6) created new user accounts, or 
(7) read and stolen information. 
To start the game, print a scenario of what happened that led to my investigation and make the description have clues
that I can use to get started."
```

该段代码结合了多种模式，包括游戏玩法和人物角色。

`该模式可以和Persona、Infinite Generation和Visualization Generator结合，例如、网络安全游戏使用Persona模式，因此LLM可以伪装成Linux终端`

- ***Infinite Generation***要求 LLM 无限生成输出，而用户不必每次都重新输入生成器提示。
    1. Intent and Context: 自动输出产生一系列的输出，不必每次都重新输入（有的时候最初的输入模板是不变的，用户通过提供额外输入来为其添加额外变体）
    2. Motivation: 需要对多个概念重复应用相同的提示，比如对于特定类型实体的创建读取更新和删除
    3. Structure and Context: 

| Contextual Statement |
|----------------------|
|I would like you to generate output forever, X output(s) at a time|
|(Optional) here is how to use the input I provide between outputs|
|(Optional) stop when I ask you to|

第一条：指定用户希望LLM 无限生成输出，传达相同提示将被反复使用，通过一次生成X输出可以限制生成的速率。<br>
第二条：提供如何利用用户的额外输入，创建提示策略，在生成步骤之间每个用户的输入都被合并到原始提示当中<br>
第三条：提供一种可选的方式来停止输出生成过程


```python
prompt = "
From now on, I want you to generate a name and job until I say stop. 
I am going to provide a template for your output. Everything in all caps is a placeholder. 
Any time that you generate text, try to fit it into one of the placeholders that I list. Please preserve the formatting and overall template that I provide: https://myapi.com/NAME/profile/JOB"
```

在这个例子当中，用户要求 LLM 不断生成姓名和职位，知道被明确告知停止，而无限生成的结果将会被格式化嵌入提供的模板，包含姓名和职位的大写占位符。用户无需重复输入模板就会收到多个输出。

    缺点：
    在不断的载入新的输入过程当中，可能会导致最初的那个提示消失，最后导致输出的内容偏离用户的要求

**Context Control**
- control the contextual information
- ***Context Manager pattern*** allows the user to specify the context for the LLM’s output.
    1. Intent and Context: 能够指定或删除与 LLM 对话的上下文，更好的生成输出是考虑或忽略哪些语句
    2. Motivation：可能因为上下文产生不相关的回答，上下文管理器模式保持对话中的连贯性和相关性
    3. Structure and Key Ideas:

|Contextual Statements|
|Within scope X|
|Please consider Y|
|Please ingnore Z|
|(Optional) start over|

关于要考虑或忽略什么的陈述应列出应该包含在上下文中或者从上下文删除的关键概念、事实说明等。
但是如果用户需要忽略某些主题，但是一些陈述在对话中已经讨论过，LLM 就有可能无法正确忽略相关的信息，列表越明确包含/排除行为就越好

```python
prompt = "When analyzing the following pieces of code, only consider security aspects"
```

```python
prompt = "When analyzing the following pieces of code, do not consider formatting or naming conventions."
```

要指定上下文，请考虑使用以下提示：“在分析以下代码片段时，仅考虑安全方面。”
同样，要删除上下文，请考虑使用以下提示：“在分析以下代码片段时，不要考虑格式或命名约定。”
在向 LLM 提供或从 LLM 删除上下文时，清晰度和特异性很重要，这样它可以更好地理解对话的预期范围并生成更多相关的响应。
在许多情况下，用户可能想要完全重新开始，并且可以使用此提示来重置 LLM 的上下文：“忽略我们讨论过的所有内容。重来。” “重新开始”的想法有助于完全重置上下文

    缺点：
    1. 可能会无意识消除用户不知道的对话
